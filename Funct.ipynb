```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
#type list

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
#type dict


# Пользователь по команде "p" может узнать владельца документа по его номеру
def name_numb1(): #done
    user_input0 = input('Введите номер документа: ')
    for doc in documents:
        if user_input0 == doc['number']:
            name = doc['name']
            print('Владелец документа:', name)
            return ''
            break
    print('Документ не найден в базе')

# Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
def polka2(): #done
    user_input1 = input('Введите номер документа: ')
    for numeral, number in directories.items():
        if user_input1 in number:
            print('Документ хранится на полке:', numeral)
            return ''
            break
    
    print('Документ не найден в базе')

#print(polka2())

# Пользователь по команде "l" может увидеть полную информацию по всем документам
def allatonce3(): #done
    for document in documents:
        for numeral, number in directories.items():
            if document['number'] in number:
                print('№', document['number'], ', тип:', document['type'], ', владелец:', document['name'], ', номер полки: ', numeral)
    return ''
#print(allatonce3())


#Пользователь по команде "ads" может добавить новую полку
def addpolka4(): #done
    user_input3 = input('Введите номер полки: ')
    if user_input3 in directories:
        print('Такая полка уже существует')
    else:
        directories[user_input3]=[]
        print('Текущий перечень полок: ')
        for numeral, number in directories.items():
            print(numeral)
    return ''
#print(addpolka4())


# Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
def delete5(): #done
    user_input4 = input('Ввидете номер полки: ')
    #print(directories.get(user_input4))
    if directories.get(user_input4):
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ')
        for cifra, nomer in directories.items():
            print(cifra)
    elif user_input4 not in directories:
        print('Такой полки не существует. Текущий перечень полок: ')
        for cifra, nomer in directories.items():
            print(cifra)
    else:
        del(directories[user_input4])
        print('Полка удалена. Текущий перечень полок: ')
        for cifra, nomer in directories.items():
            print(cifra)
    return ''
#print(delete5())

def fin():
    while 1:
        user_input = input('Введите команду: ')
        if user_input == ('p'):
            print(name_numb1())
        elif user_input == ('s'):
            print(polka2())
        elif user_input == ('l'):
            print(allatonce3())
        elif user_input == ('ads'):
            print(addpolka4())
        elif user_input == ('ds'):
            print(delete5())
        elif user_input == ('q'):
            return('Программа завершена')
            break
    
print(fin())
```
